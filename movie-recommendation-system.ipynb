```python
import ast
def clean_genre(col):
    l = []
    for i in ast.literal_eval(col):
        l.append(i['name'])
    return l
```

```python
combined_df['genres'] =combined_df['genres'].apply(clean_genre)
```

```python
combined_df.head()
```

```python
combined_df['release_year'] = pd.to_datetime(combined_df['release_date']).dt.year
```

```python
def clean_cast(col):
    l = []
    c = 0
    for i in ast.literal_eval(col):
            if c!= 3:
                l.append(i['name'])
            else:
                break
            c+=1
        
    return l
        
    
```

```python
 combined_df['cast'] = combined_df['cast'].apply(clean_cast)
```

```python
def fetch_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job']=="Director":
            l.append(i['name'])
            break
    return l
```

```python
combined_df.columns
```

```python
combined_df['crew'] = combined_df['crew'].apply(fetch_director)
```

```python
combined_df.head()
```

```python
combined_df['overview'] = combined_df['overview'].apply(lambda x:x.split())
```

```python
combined_df.head()
```

```python
combined_df['genres'] = combined_df['genres'].apply(lambda x: [i.replace(" ","") for i in x])
combined_df['cast'] = combined_df['cast'].apply(lambda x: [i.replace(" ","") for i in x])
combined_df['crew'] = combined_df['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
combined_df.head()
```

```python
combined_df['tags'] = combined_df['genres']+combined_df['overview']+combined_df['cast']+combined_df['crew']
```

```python
df = combined_df[['id','title','tags','release_year']]
```

```python
df = df.drop_duplicates(subset='id', keep='first')
```

```python
df.loc[:,'tags'] = df['tags'].apply(lambda x: " ".join(x))
```

```python
df.loc[:,'tags'] = df['tags']+' ' +combined_df['title']+' '+ df['release_year'].astype(str) 
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df.loc[:,'tags'] = df['tags'].apply(stem)
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 9000,stop_words='english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(movie):
    # first find the index of the given movie
    index = df[df['title']==movie].index[0]
    distances = similarity[index]
    movies_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(df.iloc[i[0]])
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
l = recommend('Captain America')
```
